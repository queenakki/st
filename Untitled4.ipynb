# **Outerline direction**

```python
def detect_outliers_zscore(data):
    outliers = []
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    #print(mean, std)
    for i in data:
        #print(i)
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code
```

```python
sample_outliers_tata = detect_outliers_zscore(stock['tata-volume'])
sample_outliers_Reliance_volume = detect_outliers_zscore(stock['Reliance-Volume'])
sample_outliers_amazon_volume = detect_outliers_zscore(stock['Reliance-Volume.1'])
print("Outliers in tata-volume column from Z-scores method: ", sample_outliers_tata)
print("Outliers in reliance-volume from Z-scores method: ", sample_outliers_Reliance_volume)
print("Outliers in amazon-volume column from Z-scores method: ", sample_outliers_amazon_volume)
```
